# 3. Operators

#### Arithmetic Operators

```python
a = 10
b = 3

print(a + b) # addition
print(a - b) # subtraction
print(a * b) # multiplication
print(a / b) # division
print(a // b) # floor division
print(a % b) # modulus
print(a ** b) # exponentiation
```

#### Comparison Operators

```python
a = 10
b = 20

print(a == b) # equality
print(a != b) # inequality
print(a > b) # greater than
print(a < b) # less than
print(a >= b) # greater than or equal to
print(a <= b) # less than or equal to
```

#### Logical Operators

```python
x = True
y = False

print(x and y) # logical AND
print(x or y) # logical OR
print(not x) # logical NOT
```

#### Assignment Operators

```python
x = 10

x += 5 # addition assignment
x -= 2 # subtraction assignment
x *= 3 # multiplication assignment
x /= 2 # division assignment
```

# 5. Conditional Statements

#### if Statement

```python
age = 20

if age >= 18:
    print("Eligible to Vote")
```

#### if-else Statement

```python
age = 16

if age >= 18:
    print("Adult")
else:
    print("Minor")
```

#### if-elif-else Statement

```python
marks = 85

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

# 6. Loops & Control Statementss

#### For Loop

```python
for i in range(5):
    print(i)
```

#### While Loop

```python
i = 1

while i <= 5:
    print(i)
    i += 1
```

#### Break Statement

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

#### Continue Statement

```python
for i in range(5):
    if i == 2:
        continue
    print(i)
```

# 7. Lists

#### Creating a List

```python
fruits = ["Apple", "Banana", "Mango"]
```

#### Accessing Elements

```python
print(fruits[0])
print(fruits[-1])
```

#### List Methods

```python
fruits.append("Orange")
fruits.insert(1, "Grapes")
fruits.remove("Banana")
print(len(fruits))
print(fruits)
```

#### Loop Through List

```python
for fruit in fruits:
    print(fruit)
```
