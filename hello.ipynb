# 3. Operators

#### 3.1 Arithmetic Operators

```python
a = 10
b = 3

print(a + b) # addition
print(a - b) # subtraction
print(a * b) # multiplication
print(a / b) # division
print(a // b) # floor division
print(a % b) # modulus
print(a ** b) # exponentiation
```

#### 3.2 Comparison Operators

```python
a = 10
b = 20

print(a == b) # equality
print(a != b) # inequality
print(a > b) # greater than
print(a < b) # less than
print(a >= b) # greater than or equal to
print(a <= b) # less than or equal to
```

#### 3.3 Logical Operators

```python
x = True
y = False

print(x and y) # logical AND
print(x or y) # logical OR
print(not x) # logical NOT
```

#### 3.4 Assignment Operators

```python
x = 10

x += 5 # addition assignment
x -= 2 # subtraction assignment
x *= 3 # multiplication assignment
x /= 2 # division assignment
```

# 5. Conditional Statements

#### 5.1 if Statement

```python
age = 20

if age >= 18:
    print("Eligible to Vote")
```

#### 5.2 if-else Statement

```python
age = 16

if age >= 18:
    print("Adult")
else:
    print("Minor")
```

#### 5.3 if-elif-else Statement

```python
marks = 85

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

# 6. Loops & Control Statementss

#### 6.1 For Loop

```python
for i in range(5):
    print(i)
```

#### 6.2 While Loop

```python
i = 1

while i <= 5:
    print(i)
    i += 1
```

#### 6.3 Break Statement

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

#### 6.4 Continue Statement

```python
for i in range(5):
    if i == 2:
        continue
    print(i)
```

# 11. Functions

#### 11.1 Function Without Parameters

```python
def greet():
    print("Welcome to Python")
greet()
```

#### 11.2 Function With Parameters

```python
def greet(name):
    print("Hello", name)
greet("Nemis")
```

#### 11.3 Function With Return Value

```python
def add(a, b):
    return a + b
result = add(10, 20)
print(result)
```

#### 11.4 Default Parameter

```python
def greet(name="Guest"):
    print("Hello", name)
greet()
greet("Nemis")
```

#### 11.5 Lambda Function

```python
square = lambda x: x * x
print(square(5))
```
